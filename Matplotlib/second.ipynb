### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
# scatter plot -> axis level function
tips=sns.load_dataset('tips')
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

#### Line Plot

```python
gap=px.data.gapminder()
gap
```

```python
gap_for_india=gap[gap['country']=='India']
gap_for_india
```

```python
# axes level function
sns.lineplot(data=gap_for_india,x='year',y='lifeExp')
```

```python
# figure level
sns.relplot(data=gap_for_india,x='year',y='lifeExp',kind='line')
```

```python
gap_for_southasia=gap[gap['country'].isin(['India','Pakistan','China'])]
# gap_for_southasia
sns.relplot(data=gap_for_southasia,x='year',y='lifeExp',kind='line',hue='country')
```

### Facet Plot

```python
# only work with figure level function
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='smoker',row='sex')
```

```python
# col wrap
sns.relplot(data=gap,x='year',y='lifeExp',kind='line',row='continent')
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# plotting univariate histogram
# figure level -> displot
# axes level -> histplot -> kdeplot ->rugplot
```

```python
# plotting univariate histogram
tips
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# bins parameter 
sns.displot(data=tips,x='total_bill',kind='hist',bins=2)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

sns.displot(data=tips,x='day',kind='hist',bins=2)
```

```python
# hue parameter
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',hue='sex',element='step')
```

```python
# faciting using rows and columns
sns.displot(data=titanic,x='age',hue='sex',col='survived')
```

### KDE Plot

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

### Rug Plot

```python
# Rugplot
# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
# sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap
# Plot rectangular data as a color-encoded matrix
country_life_expectancy_per_year=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
sns.heatmap(data=country_life_expectancy_per_year)
```

```python
# annot
europians_life_expectancy=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(data=europians_life_expectancy,annot=True,linewidths=0.5)
```

```python
# Clustermap

iris_data=px.data.iris()
iris_data
```

```python
sns.clustermap(iris_data.iloc[:,[0,1,2,3]])
```
