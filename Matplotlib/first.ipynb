```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
price=[45000,54000,57000,49000,47000]
year=[2015,2016,2017,2019,2020]
plt.plot(year,price)
```

```python
# creating pandas dataframe
batsman=pd.read_csv("./sharma-kohli.csv")
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# giving labels and Title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma Vs Virat Kohli IPL Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scrored")
```

```python
#giving colors by name, hex code and styling lines
# giving labels and Title
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dashed",linewidth=3) #other linestyle avaliable like "dotted","dashdot",
plt.plot(batsman['index'],batsman['RG Sharma'],color="blue")
plt.title("Rohit Sharma Vs Virat Kohli IPL Career Comparison")
plt.xlabel("Seasons")
plt.ylabel("Runs Scrored")
```

```python
# legends and markers
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashdot',linewidth=3,marker='o',label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker=">",markersize=10,label="Sharma")
plt.title('Sharma Vs Kohli (IPL Runs) Comparion')
plt.ylabel('Runs Scored')
plt.xlabel('Seasons')
plt.legend(loc='best')# by default location is choosed where more space is avalaibe
```

```python
# limiting the axis
plt.plot(year,price)
plt.xlim(2016,2019)
plt.ylim(top=55000)
```

## Bar Chart
###### .Bivariate Analysis
###### .Numerical vs Categorical Analysis
###### .Use case - Aggregate Analysis of groups

```python
 # simple bar chart
children=[10,20,40,10,30]
colors=['red','blue','green','yellow','pink']

plt.bar(colors,children)
```

```python
# bar chart using data
```

```python
# horizontal bar chart
plt.barh(colors,children)
```

```python
# colr and label
batsman_season_records=pd.read_csv('./batsman_season_record.csv')
batsman_season_records
```

```python
# creating bars side by side
# Note: There is no direct function to achieve this goal, but we can achieve this by manipulating the positions of bars
plt.bar(np.arange(batsman_season_records.shape[0])-0.2,batsman_season_records['2015'],width=0.2)
plt.bar(np.arange(batsman_season_records.shape[0]),batsman_season_records['2016'],width=0.2)
plt.bar(np.arange(batsman_season_records.shape[0])+0.2,batsman_season_records['2017'],width=0.2)

plt.xticks(np.arange(batsman_season_records.shape[0]),batsman_season_records['batsman'])
plt.show()
```

```python
# a problem: Lets you have high cardinality data or categorial data with larger names, then labels overlap
# Solution it can be solve by using plt.xticks() 
plt.bar(colors,children)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# stacked bar charts
plt.bar(batsman_season_records['batsman'],batsman_season_records['2017'],label='2017')+
plt.bar(batsman_season_records['batsman'],batsman_season_records['2016'],bottom=batsman_season_records['2017'],label='2016')
plt.bar(batsman_season_records['batsman'],batsman_season_records['2015'],bottom=batsman_season_records['2016']+batsman_season_records['2017'],label='2015')
plt.legend()
```

### Pie Chart
###### ==>Univariate/Bivariate Analysis
###### ==>Categorical vs numerical
###### ==>Use case - To find contribution in a whole/standard scale

```python
# simple data
data=[23,45,100,20,49]
subjects=['English','Chemistry','Computer Science','Urdu','Maths']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# data set
chris_gayle=pd.read_csv('./gayle-175.csv')
plt.pie(chris_gayle['batsman_runs'],labels=chris_gayle['batsman'],autopct='%0.0f%%')
plt.show()
```

```python
# highlighting by exploding the Pie Slice
plt.pie(chris_gayle['batsman_runs'],labels=chris_gayle['batsman'],autopct='%0.1f%%',explode=[0,.1,0,0,0,0])
plt.show()
```
